```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib import cm
```

```python
data = pd.read_csv('all_seasons.csv')
print(data.shape)
```

```python
df = data.groupby(by = 'college')
cnt = df['player_name'].nunique()
cnt = cnt.sort_values(ascending = False)
print(type(cnt), len(cnt))
#print(cnt[1:21])
```

```python
colleges = cnt[1:21]
x = colleges.index
print(type(x), x)
y = colleges.values
print(type(y), y)
```

```python
i = np.arange(20)
norm = plt.Normalize(10,70)
norm_values = norm(y)
map_vir = cm.get_cmap(name='Blues')
colors = map_vir(norm_values)
fig = plt.figure(figsize = (15, 10))
plt.subplot(111)
ax = plt.barh(i, y, color=colors, tick_label = x)
plt.gca().invert_yaxis()
sm = cm.ScalarMappable(cmap=map_vir,norm=norm)
sm.set_array([])
plt.colorbar(sm)
plt.yticks(fontproperties = 'Times New Roman', size = 14)
plt.xticks(fontproperties = 'Times New Roman', size = 14)
for a, b in zip(i, y):
    plt.text(b, a, b, fontsize=10)
plt.xlabel('Number of NBA Players', fontdict={'family' : 'Times New Roman', 'size'   : 16})
plt.ylabel('College', fontdict={'family' : 'Times New Roman', 'size'   : 16})
plt.show()
```

```python
players = data.groupby('player_name')['season'].count()
players = players[players >= 5]
print(type(players), len(players))
print(type(players.index))
cnt_players = data.loc[data['player_name'].isin(players.index)]
print(data.shape)
print(cnt_players.shape, type(cnt_players))
```

```python
def wavg(group, avg_name, weight_name):
    d = group[avg_name]
    w = group[weight_name]
    try:
        return (d * w).sum() / w.sum()
    except ZeroDivisionError:
        return d.mean()

nts = cnt_players.groupby('player_name').apply(wavg, 'pts', 'gp').sort_values(ascending = False)
#players = data.groupby(by = 'player_name')
#players = players['ts_pct'].sum()
#players = players.sort_values(ascending = False)
print(nts.head(10))
```

```python
x1 = nts.index
y1 = nts.values
print(len(x1), len(y1))
y2 = []
y3 = []
y4 = []
for i in x1:
    current = data[data['player_name'] == i]
    pct = wavg(current, 'ts_pct', 'gp')
    y2.append(pct)
    ts = wavg(current, 'net_rating', 'gp')
    y3.append(ts)
    #y3.append(data[data['player_name'] == i]['season'].count())
    usg = wavg(current, 'usg_pct', 'gp')
    y4.append(usg)
    
print(max(y1[:100]), min(y1[:200]))
print(len(y2), y2[:10], max(y2[:200]), min(y2[:200]))
print(len(y3), y3[:10], max(y3[:200]), min(y3[:200]))
print(len(y4), y4[:10], max(y4[:200]), min(y4[:200]))
```

```python
i = 100
norm = plt.Normalize(0.17, 0.32)
norm_values = norm(y4[:i])
map_vir = cm.get_cmap(name='BuPu')
colors = map_vir(norm_values)
fig = plt.figure(figsize = (36, 27))
plt.subplot(111)
norms = plt.Normalize(-9.4, 10.22)
size_values = norms(y3[:i])
ax = plt.scatter(y1[:i], y2[:i], s = [5000 * i for i in size_values], color = colors, alpha = 0.5)
sm = cm.ScalarMappable(cmap=map_vir,norm=norm)
sm.set_array([])
cb = plt.colorbar(sm)
cb.set_label('Average Usage Percentage', fontdict = {'family' : 'Times New Roman', 'size'   : 20})
for a, b, c in zip(x1[:i], y1[:i], y2[:i]):
    plt.text(b, c, a, fontdict = {'family' : 'Times New Roman', 'size'   : 16})
plt.yticks(fontproperties = 'Times New Roman', size = 16)
plt.xticks(fontproperties = 'Times New Roman', size = 16)
plt.xlabel('Average Points', fontdict = {'family' : 'Times New Roman', 'size'   : 20})
plt.ylabel('Average True Percentage', fontdict = {'family' : 'Times New Roman', 'size'   : 20})
plt.savefig('nba_players.png')
plt.show()
```
